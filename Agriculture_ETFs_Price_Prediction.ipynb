## 1. Data and Library Importation

```python
# Libraries

!pip install --upgrade yfinance

from sklearn.linear_model import LinearRegression

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

import yfinance as yf

#import pandas_datareader.data as web
```

```python
# Import Data

sb_data = yf.download('SOYB', '2011-09-23', '2021-09-23', auto_adjust=True)
```

```python
# Test

sb_data.head()
```

```python
# Keeping only close value column

#---

sb_data = sb_data[['Close']]

sb_data.head()

#---
```

```python
# Removing missing rows if applicable

sb_data = sb_data.dropna()

# Plotting closing price of the SOYB ETF

sb_data.Close.plot(figsize=(10,7), color='r')
plt.ylabel('Soybean ETF Prices')
plt.title('Soybean ETF Price Series')
plt.show()
```

## 2. Data preprocessing

* Determination of dependent and independent variables
* Train-Test split 

```python
sb_data['S_3'] = sb_data['Close'].rolling(window=3).mean()
sb_data['S_9'] = sb_data['Close'].rolling(window=9).mean()
sb_data['next_day_price'] = sb_data['Close'].shift(-1)

sb_data = sb_data.dropna()
X = sb_data[['S_3', 'S_9']]

y = sb_data['next_day_price']
```

```python
y.head()
```

```python
sb_data.head()
```

```python
# Train-test split

t = .8
t = int(t*len(sb_data))

X_train = X[:t]
y_train = y[:t]

X_test = X[t:]
y_test = y[t:]
```

# 3. Creating a Linear Regression Model

```python
linear = LinearRegression().fit(X_train, y_train)
print('Linear Regression Model')
print('Soybean ETF Price: \n (y) = %.2f * 3 Days Moving Average (x1) \
+ %.2f * 9 Days Moving Average (x2) \
+ %.2f (constant)' % (linear.coef_[0], linear.coef_[1], linear.intercept_))
```

```python
# Predicting Prices

predicted_price = linear.predict(X_test)

predicted_price = pd.DataFrame(
    
    predicted_price,
    index = y_test.index, 
    columns = ['price']

)

predicted_price.plot(figsize=(10,7))

y_test.plot()

plt.title('Soybean Forecast vs Actual - October 2019 to September 2021')

plt.legend(['Predicted Price', 'Actual Price'])

plt.ylabel('Soybean ETF Price')

plt.show()
```

```python
# Computing R^2

r2_score = linear.score(X[t:], y[t:])*100
float('{0:.2f}'.format(r2_score))
```

### Calculating Cumulative Results

```python
sb2_data = pd.DataFrame()

sb2_data['price'] = sb_data[t:]['Close']
sb2_data['predicted_price_next_day'] = predicted_price
sb2_data['actual_price_next_day'] = y_test
sb2_data['soybean_returns'] = sb2_data['price'].pct_change().shift(-1)

sb2_data['signal'] = np.where(sb2_data.predicted_price_next_day.shift(1) < sb2_data.predicted_price_next_day, 1, 0)

sb2_data['strategy_returns'] = sb2_data.signal * sb2_data['soybean_returns']

((sb2_data['strategy_returns']+1).cumprod()).plot(figsize=(10,7),color='g')

plt.ylabel('Cumulative Returns')

plt.show()
```

## Calculating the Sharpe Ratio

```python
sb_sharpe = sb2_data['strategy_returns'].mean() / sb2_data['strategy_returns'].std()*(252**0.5)

'Soybeans Sharpe Ratio %.2f' % (sb_sharpe)
```

## Predicting Daily Moves

```python
import datetime as dt

current_date = dt.datetime.now()

current_date
```

```python
data = yf.download('SOYB', '2011-09-23', current_date, auto_adjust=True)
data['S_3'] = data['Close'].rolling(window=3).mean()
data['S_9'] = data['Close'].rolling(window=9).mean()
data = data.dropna()

# Forecasting price
data['predicted_soybean_price'] = linear.predict(data[['S_3', 'S_9']])
data['signal'] = np.where(data.predicted_soybean_price.shift(1) < data.predicted_soybean_price, 'Buy', 'No Position')

data.tail(10)[['signal', 'predicted_soybean_price']]
```
